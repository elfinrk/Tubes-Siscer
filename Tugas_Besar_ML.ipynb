#### 1.1 Latar Belakang
Stres merupakan salah satu permasalahan psikologis yang dapat memengaruhi kesehatan mental, kondisi fisik, serta produktivitas individu. Menurut World Health Organization (WHO), kesehatan mental dipengaruhi oleh berbagai faktor individu, sosial, dan lingkungan yang saling berhubungan sehingga diperlukan perhatian terhadap faktor-faktor yang dapat memengaruhinya [1]. Selain itu, penelitian menunjukkan bahwa berbagai tekanan yang berasal dari aspek akademik, sosial, maupun personal dapat menjadi faktor yang berkontribusi terhadap meningkatnya kondisi stres pada individu [2]. Kondisi tersebut menunjukkan bahwa diperlukan pendekatan yang mampu mengidentifikasi pola faktor-faktor yang berhubungan dengan tingkat stres secara lebih sistematis, sehingga proses klasifikasi tingkat stres dapat dilakukan secara lebih objektif berdasarkan karakteristik data individu. 

Mahasiswa merupakan salah satu kelompok yang rentan mengalami stres karena menghadapi berbagai tuntutan selama masa perkuliahan. Beban akademik, kondisi psikologis, tekanan pencapaian prestasi, faktor sosial, serta gaya hidup dapat berhubungan dengan tingkat stres yang dialami mahasiswa [2], [3]. Tingkat stres yang tinggi dapat berdampak pada kesehatan mental, kualitas hidup, dan performa akademik mahasiswa. Namun, perbedaan karakteristik data dan pola hubungan antar faktor yang memengaruhi stres dapat menyebabkan setiap algoritma Machine Learning memiliki kemampuan klasifikasi yang berbeda. Oleh karena itu, diperlukan evaluasi terhadap beberapa algoritma klasifikasi untuk mengetahui pendekatan yang mampu memberikan performa terbaik dalam mengidentifikasi tingkat stres mahasiswa. 

Perkembangan teknologi Machine Learning memberikan peluang dalam melakukan klasifikasi tingkat stres dengan memanfaatkan pola hubungan antar variabel pada dataset. Penelitian ini menggunakan dataset Student Stress Factors: A Comprehensive Analysis yang mencakup faktor akademik, psikologis, sosial, fisiologis, dan lingkungan, sehingga diperlukan proses seleksi fitur untuk menentukan atribut yang memiliki hubungan paling relevan terhadap tingkat stres [4]. Untuk memperoleh fitur yang lebih relevan terhadap proses klasifikasi, dilakukan seleksi fitur menggunakan korelasi Pearson dengan memilih variabel yang memiliki nilai korelasi absolut |r| ≥ 0,6 terhadap variabel target stress_level. Pemilihan batas tersebut didasarkan pada interpretasi bahwa nilai korelasi yang semakin mendekati -1 atau 1 menunjukkan hubungan yang semakin kuat antar variabel [5]. Penggunaan nilai absolut dilakukan agar fitur dengan hubungan negatif tetap dipertahankan karena tetap menunjukkan hubungan yang kuat terhadap stress_level, meskipun memiliki arah hubungan yang berlawanan. Penggunaan F1-Score dipilih karena mampu mempertimbangkan keseimbangan antara precision dan recall dalam proses klasifikasi multi-kelas tingkat stres.

#### Deteksi Pencilan Menggunakan Visualisasi Boxplot

`Konteks` : Cell ini kami gunakan untuk mendeteksi keberadaan pencilan (outliers) secara menyeluruh pada 20 fitur input di dalam dataset melalui grafik Box and Whisker Plot (Boxplot). Melalui visualisasi matriks sub-grafik ini, kami dapat mengamati karakteristik persebaran nilai dari seluruh dimensi indikator mahasiswa berdasarkan kelas target stress_level secara instan.

`plt.figure(figsize=(20, 22))` : Perintah ini kami gunakan untuk menyiapkan bingkai kanvas grafik baru dengan ukuran lebar 20 inci dan tinggi 22 inci agar seluruh 20 fitur dapat tertampung secara rapi dan proporsional.

`for i, col in enumerate(features, 1)` : Perintah perulangan (looping) ini kami gunakan untuk mengotomatisasi pembuatan grafik boxplot bagi seluruh fitur input satu per satu tanpa perlu menulis kode yang sama sebanyak 20 kali.

`plt.subplot(5, 4, i)` : Perintah ini kami gunakan untuk membagi kanvas besar menjadi matriks sub-grafik kecil dengan tata letak 5 baris dan 4 kolom, pas sesuai dengan jumlah 20 fitur input kami.

`sns.boxplot(x='stress_level', y=col, data=df, hue='stress_level', palette='Set2', legend=False)` : Perintah utama dari library Seaborn yang kami gunakan untuk menggambar bentuk boxplot dari setiap fitur (y=col) terhadap kelas tingkat stres (x='stress_level'). Kami menambahkan parameter hue dan legend=False untuk menyelaraskan pewarnaan sekaligus menghindari pesan peringatan (FutureWarning) pada sistem.

`plt.title(...), plt.xlabel(...), plt.ylabel(...)` : Perintah-perintah ini kami gunakan untuk memberikan keterangan judul sub-grafik serta label penanda pada sumbu X dan sumbu Y agar mudah dibaca oleh dosen.

`plt.tight_layout()` : Perintah ini kami gunakan untuk menyesuaikan jarak dan posisi antar-grafik secara otomatis agar susunannya rapi dan teks tidak saling bertumpukan.

`plt.show()` : Perintah terakhir ini kami gunakan untuk memproses seluruh rangkaian kode di atas dan langsung menampilkan visualisasi 20 boxplot tersebut ke layar.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat data tingkat stres mahasiswa
df = pd.read_csv("StressLevelDataset.csv")

# Mengisolasi daftar 20 fitur input dengan menghapus kolom target
features = df.drop(columns=['stress_level']).columns.tolist()

# Menyiapkan dimensi kanvas besar untuk menampung seluruh fitur input
plt.figure(figsize=(20, 22))

# Melakukan perulangan otomatis untuk menggambar boxplot dari ke-20 fitur input
for i, col in enumerate(features, 1):
    plt.subplot(5, 4, i)
    # Menambahkan hue dan legend=False agar bebas dari FutureWarning
    sns.boxplot(x='stress_level', y=col, data=df, hue='stress_level', palette='Set2', legend=False)
    plt.title(f'Distribusi {col}', fontsize=10, fontweight='bold')
    plt.xlabel('Stress Level')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

#### Analisis Data Pencilan (Outlier Analysis)

`Konteks` : Bagian ini kami gunakan untuk mendeteksi dan menghitung keberadaan data pencilan (outliers) pada seluruh 20 fitur input secara matematis. Melalui analisis ini, kami ingin memeriksa kualitas keaslian data responden pada setiap kolom secara terbuka sebelum menentukan langkah pra-pemrosesan data atau masuk ke tahap pemodelan machine learning.

`Q1 = df_features.quantile(0.25)` : Perintah ini kami gunakan untuk menghitung nilai Kuartil 1 (25% data terbawah) di setiap kolom untuk mencari batas bawah data wajar.

`Q3 = df_features.quantile(0.75)` : Perintah ini kami gunakan untuk menghitung nilai Kuartil 3 (75% data teratas) di setiap kolom untuk mencari batas atas data wajar.

`IQR = Q3 - Q1` : Perintah ini kami gunakan untuk menghitung selisih jarak antara Kuartil 3 dan Kuartil 1 guna menentukan rentang penyebaran data utama.

`outliers = ((df_features < ... ) | (df_features > ...)).sum()` : Rumus statistik standar (Metode IQR) yang kami gunakan untuk menghitung total jumlah data pencilan yang berada di luar batas bawah maupun batas atas secara otomatis.

`print(outliers)` : Perintah terakhir ini kami gunakan untuk mencetak dan memunculkan hasil deteksi pencilan dari seluruh 20 fitur secara utuh tanpa ada yang disembunyikan.

```python
import pandas as pd

# Memuat data tingkat stres mahasiswa
df = pd.read_csv("StressLevelDataset.csv")

# Mengisolasi data hanya pada 20 fitur input
df_features = df.drop(columns=['stress_level'])

# Menghitung batas kuartil dan Interquartile Range (IQR)
Q1 = df_features.quantile(0.25)
Q3 = df_features.quantile(0.75)
IQR = Q3 - Q1

# Mendeteksi jumlah pencilan di setiap kolom fitur
outliers = ((df_features < (Q1 - 1.5 * IQR)) | (df_features > (Q3 + 1.5 * IQR))).sum()

# Menampilkan hasil deteksi pencilan untuk SELURUH 20 fitur input
print("Hasil Deteksi Pencilan Lengkap Seluruh Fitur:")
print(outliers)
```

#### 3.1 Feature Selection (Seleksi Fitur Berbasis Filter Method)

`Konteks` : Bagian ini kami gunakan untuk melakukan seleksi fitur menggunakan metode penyaringan (Filter Method) berbasis nilai Korelasi Pearson. Demi memenuhi validitas ilmiah yang kuat di hadapan penguji, kami melakukan eksperimen perbandingan terhadap tiga nilai ambang batas (threshold) yang berbeda (0.4, 0.6, dan 0.7) untuk menganalisis penyusutan dimensi data secara objektif sebelum mengunci keputusan final pada angka yang paling optimal.

`for t in [0.4, 0.6, 0.7]:` : Perintah perulangan untuk mensimulasikan tiga skenario ambang batas korelasi (longgar, ketat, dan sangat ketat) sebagai bahan perbandingan keputusan kelompok kami.

`abs(korelasi_target) >= threshold_final` : Mengambil nilai mutlak korelasi (mengabaikan arah hubungan positif/negatif) untuk menyaring fitur secara otomatis berdasarkan keputusan final kelompok.

`df_selected = df[...] `: Membuat DataFrame final yang hanya meloloskan fitur-fitur terbaik hasil seleksi akhir kelompok kami beserta targetnya.

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif

# 1. Memuat data tingkat stres mahasiswa
df = pd.read_csv("StressLevelDataset.csv")
X = df.drop(columns=['stress_level'])
y = df['stress_level']

# 2. Menghitung Korelasi Pearson (Nilai Mutlak)
pearson_scores = X.apply(lambda col: abs(df['stress_level'].corr(col)))

# 3. Menghitung Mutual Information (Kemampuan mendeteksi Non-Linear)
# Menggunakan random_state=42 agar hasilnya konsisten saat di-running dosen
mi_scores = mutual_info_classif(X, y, random_state=42)
mi_series = pd.Series(mi_scores, index=X.columns)

# 4. Membuat DataFrame Perbandingan Metode Seleksi Fitur
df_compare = pd.DataFrame({
    'Pearson (Linear)': pearson_scores,
    'Mutual Information (Non-Linear)': mi_series
}).sort_values(by='Pearson (Linear)', ascending=False)

print("TABEL EKSPERIMEN PERBANDINGAN METODE SELEKSI FITUR:")
print("="*70)
print(df_compare.round(4))
print("="*70)
```

Sebagai solusinya, eksperimen ini dijalankan menggunakan Stratified K-Fold Cross-Validation (5 lipatan data). Di setiap lipatan data, kami menguji seluruh variasi parameter menggunakan GridSearchCV dengan fokus penilaian pada metrik F1-Score Macro agar penilaian performa adil di ketiga tingkatan kelas stres.

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

print("=== MEMULAI PROSES TUNING SEMUA PARAMETER KNN ===")

# 1. Mengubah DataFrame menjadi array agar mudah diindeks oleh K-Fold manual
X_arr = X.values
y_arr = y.values

# 2. Bagi data menjadi 5 lipatan secara adil sesuai aturan panduan
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 3. Daftar variasi parameter KNN yang wajib dicoba untuk dibandingkan
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11, 13, 15],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

# Tempat menyimpan skor performa dari setiap lipatan data
fold_scores = []

# 4. Perulangan K-Fold Manual untuk Menghindari Data Leakage Tanpa Pipeline
for fold, (train_index, test_index) in enumerate(skf.split(X_arr, y_arr), 1):
    # Memisahkan data latih dan data uji untuk fold ini
    X_train, X_test = X_arr[train_index], X_arr[test_index]
    y_train, y_test = y_arr[train_index], y_arr[test_index]
    
    # PROSES SCALING MANUAL PER FOLD (Aman dari Data Leakage)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train) # Menghitung skala hanya dari data latih fold ini
    X_test_scaled = scaler.transform(X_test)       # Mengubah data uji menggunakan skala data latih
    
    # Menjalankan GridSearchCV dasar di dalam fold
    knn_base = KNeighborsClassifier()
    grid_search = GridSearchCV(estimator=knn_base, param_grid=param_grid_knn, scoring='f1_macro', n_jobs=-1, cv=3)
    grid_search.fit(X_train_scaled, y_train)
    
    # Menguji model terbaik hasil tuning pada data uji fold ini yang steril
    best_model_fold = grid_search.best_estimator_
    skor_fold = grid_search.score(X_test_scaled, y_test)
    fold_scores.append(skor_fold)
    
    print(f"Fold {fold} selesai. Parameter Terbaik: {grid_search.best_params_} | Skor F1: {skor_fold*100:.2f}%")

# 5. Menghitung rata-rata F1-Score dari 5 kali pengujian
rata_rata_f1 = np.mean(fold_scores) * 100

print("\n=== HASIL AKHIR EVALUASI KNN ===")
print("Rata-Rata Skor F1-Macro Validasi (5-Fold): {:.2f}%".format(rata_rata_f1))
```

#### **Penjelasan Komponen Kode dan Teori**
Formula Jarak KNN (Metrik Pengukuran): KNN bekerja dengan cara menghitung seberapa dekat jarak antara data mahasiswa baru dengan data mahasiswa yang sudah ada di memori. Dua rumus jarak utama yang kami adu dan bandingkan di dalam eksperimen ini adalah:

Euclidean Distance: Menghitung jarak berdasarkan garis lurus diagonal dari satu titik ke titik lain.

Manhattan Distance: Menghitung jarak berdasarkan garis tegak lurus (seperti cara kita berjalan menyusuri blok perumahan atau kota).

Stratified K-Fold (5 Lipatan Manual): Berfungsi memecah total data kita menjadi 5 bagian secara bergantian lewat perintah perulangan (for loop). Kata Stratified artinya komputer wajib menjaga porsi sebaran tingkat stres di setiap lipatan data agar tetap seimbang (sama rata), sehingga model diuji secara adil dan jujur.

GridSearchCV: Fitur ini bertugas menguji seluruh 28 kombinasi parameter secara otomatis di setiap lipatan data untuk mencari setelan mana yang menghasilkan skor F1-Macro paling tinggi.

#### **Analisis & Justifikasi Kelompok**
Alasan Angka $K$ Ganjil [3, 5, 7, 9, 11, 13, 15]: Kami memilih angka ganjil untuk menghindari kondisi suara berimbang atau seri (tie vote) saat voting tetangga pada 3 kelas target kita. Rentang dibatasi sampai 15 agar batas keputusan tidak terlalu kabur (oversmoothed) yang bisa menurunkan akurasi model.

Justifikasi Pilihan Jarak Euclidean dibanding Manhattan (Menjawab Revisi Poin 1): Berdasarkan hasil uji kros-validasi riil, model KNN mencapai performa rata-rata 86.45% dan secara konsisten memilih Euclidean Distance di kelima lipatan data (fold).


Alasan Ilmiah: Meskipun dataset ini memiliki 17 fitur, rumus Euclidean (yang menghitung jarak lurus diagonal atau garis lurus spasial) terbukti lebih unggul dalam memetakan sebaran data tingkat stres mahasiswa ini. Hal ini mengindikasikan bahwa fitur-fitur numerik pada dataset ini menyebar dalam ruang geometris yang kontinu dan lurus, di mana kombinasi perubahan nilai antar-fitur (seperti kecemasan dan jam tidur) lebih akurat jika dihitung magnitudo totalnya secara diagonal (Euclidean) daripada dihitung terpisah per sumbu tegak lurus (Manhattan).

**Analisis Kompleksitas / Complexity Analysis** :


Waktu Latihan (Training): $\mathcal{O}(1)$. KNN adalah lazy learner (algoritma malas). Saat latihan, dia tidak menghitung rumus matematika yang berat, melainkan hanya menyalin data ke memori sehingga proses latihannya berjalan instan.

Waktu Menebak (Prediction): $\mathcal{O}(N \times D)$. Proses berat terjadi saat menebak data baru. Komputer harus menghitung jarak data baru tersebut ke seluruh ($N$) 1.100 data mahasiswa di memori untuk setiap ($D$) 17 kolom fitur yang ada.

## Eksperimen dan Tuning Model Decision Tree

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

print("=== MEMULAI TUNING PARAMETER DECISION TREE ===")

# 1. Ubah data ke bentuk array agar bisa dipotong manual
X_arr, y_arr = X.values, y.values

# 2. Siapkan pembagi data 5-Fold (Wajib sesuai aturan)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 3. Daftar parameter pemangkasan pohon yang wajib dibandingkan
param_grid_dt = {
    'criterion': ['gini', 'entropy'],    
    'max_depth': [None, 5, 10, 15, 20],  
    'min_samples_split': [2, 5, 10]      
}

fold_scores = []

# 4. Perulangan K-Fold Manual untuk Mencegah Data Leakage
for fold, (train_index, test_index) in enumerate(skf.split(X_arr, y_arr), 1):
    X_train, X_test = X_arr[train_index], X_arr[test_index]
    y_train, y_test = y_arr[train_index], y_arr[test_index]
    
    # Standardisasi manual di dalam fold (Aman dari kebocoran data)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # Cari parameter terbaik menggunakan GridSearchCV dasar
    grid_search = GridSearchCV(estimator=DecisionTreeClassifier(random_state=42), param_grid=param_grid_dt, scoring='f1_macro', cv=3, n_jobs=-1)
    grid_search.fit(X_train_scaled, y_train)
    
    # Tes model terbaik pada data uji fold ini yang masih steril
    skor_fold = grid_search.score(X_test_scaled, y_test)
    fold_scores.append(skor_fold)
    
    print(f"Fold {fold} - Setelan Terbaik: {grid_search.best_params_} | F1: {skor_fold*100:.2f}%")

# 5. Hitung rata-rata performa final
rata_rata_f1 = np.mean(fold_scores) * 100
print("\n=== HASIL AKHIR EVALUASI DECISION TREE ===")
print("Rata-Rata Skor F1-Macro Validasi (5-Fold): {:.2f}%".format(rata_rata_f1))
```

Keunggulan pohon keputusan yang mampu mempelihatkan alur logika Ya/Tidak secara visual sangat membantu Dosen Wali atau Tim Konseling dalam memahami faktor dominan apa saja yang memicu stres mahasiswa.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, f1_score

print("=== MEMULAI EVALUASI ===")

# 1. Konfigurasi Awal Pembagi K-Fold Manual dan Penamaan Kelas
# (Asumsi variabel data fitur 'X' dan target 'y' sudah didefinisikan di baris kode atas)
X_arr, y_arr = X.values, y.values
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
target_names = ['Rendah (0)', 'Sedang (1)', 'Tinggi (2)']

# 2. Inisialisasi Model Menggunakan Parameter Terbaik Hasil Tuning Kelompok
model_init = {
    'Naïve Bayes': GaussianNB(var_smoothing=1e-9),               
    'K-Nearest Neighbors (KNN)': KNeighborsClassifier(metric='euclidean', n_neighbors=3, weights='uniform'), 
    'Decision Tree': DecisionTreeClassifier(criterion='entropy', max_depth=None, min_samples_split=5, random_state=42)             
}

# Tempat penyimpanan akumulasi prediksi riil dan peluang dari tiap fold steril
model_final_predictions = {}
model_final_probabilities = {}
model_f1_scores = {}

# 3. Jalankan Proses Putaran K-Fold Manual untuk Mengeliminasi Data Leakage
for nama_model, model in model_init.items():
    y_pred_total = np.zeros(len(y_arr))
    y_proba_total = np.zeros((len(y_arr), len(target_names)))
    
    # Memotong data secara bergantian di dalam loop (Kros-Validasi Murni)
    for train_index, test_index in skf.split(X_arr, y_arr):
        X_train, X_test = X_arr[train_index], X_arr[test_index]
        y_train, y_test = y_arr[train_index], y_arr[test_index]
        
        # Penskalaan data WAJIB terisolasi penuh di dalam fold latihan
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        
        # Latih model menggunakan data skala steril
        model.fit(X_train_scaled, y_train)
        
        # Simpan hasil tebakan dan peluang probabilitas dari data uji fold ini
        y_pred_total[test_index] = model.predict(X_test_scaled)
        y_proba_total[test_index] = model.predict_proba(X_test_scaled)
        
    # Rekam hasil akumulasi keseluruhan putaran fold ke dalam dictionary
    model_final_predictions[nama_model] = y_pred_total
    model_final_probabilities[nama_model] = y_proba_total
    
    # Hitung Juri Tunggal F1-Macro Akhir
    model_f1_scores[nama_model] = f1_score(y_arr, y_pred_total, average='macro')

# 4. Urutkan Daftar Model Berdasarkan Nilai F1-Score Tertinggi (Descending)
daftar_model_urut = sorted(model_init.items(), key=lambda x: model_f1_scores[x[0]], reverse=True)

# 5. Tampilkan Output Cetakan Laporan Klasifikasi dan Visualisasi Grafik
for nama_model, model in daftar_model_urut:
    y_pred = model_final_predictions[nama_model]
    y_proba = model_final_probabilities[nama_model]
    
    print("\n" + "="*60)
    print(f" HASIL EVALUASI CROSS-VALIDATION MURNI (URUT F1-SCORE TERTINGGI): {nama_model}")
    print(f" [Macro F1-Score: {model_f1_scores[nama_model]*100:.2f}%]")
    print("="*60)
    
    # Hitung nilai rata-rata tingkat keyakinan (Confidence Score) dari fungsi probabilitas
    confidence = np.max(y_proba, axis=1)
    conf_benar = np.mean(confidence[y_pred == y_arr]) * 100
    conf_salah = np.mean(confidence[y_pred != y_arr]) * 100 if len(confidence[y_pred != y_arr]) > 0 else 0

    # Cetak laporan klasifikasi (Precision, Recall, F1 per kelas)
    print(classification_report(y_arr, y_pred, target_names=target_names))
    print(f"--> Rata-rata Confidence Prediksi BENAR : {conf_benar:.2f}%")
    print(f"--> Rata-rata Confidence Prediksi SALAH : {conf_salah:.2f}%\n")
    
    # Plot Confusion Matrix Otomatis Sesuai Aturan Skema Warna Estetik
    cm = confusion_matrix(y_arr, y_pred)
    cmap_color = 'Oranges' if nama_model == 'Naïve Bayes' else ('Blues' if nama_model == 'K-Nearest Neighbors (KNN)' else 'Greens')
    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
    disp.plot(cmap=cmap_color)
    plt.title(f"Confusion Matrix - {nama_model}")
    plt.grid(False)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree

# 1. Ambil model Decision Tree yang sudah dilatih dari kode K-Fold sebelumnya
# (Kita latih sekali lagi pada seluruh data skala steril untuk mengekstrak struktur globalnya)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled_full = scaler.fit_transform(X)

dt_model = DecisionTreeClassifier(
    criterion="entropy", max_depth=3, min_samples_split=5, random_state=42
)
dt_model.fit(X_scaled_full, y_arr)

# ============================================================
# A. FEATURE IMPORTANCE & FAKTOR DOMINAN
# ============================================================
importances = dt_model.feature_importances_
indices = np.argsort(importances)[::-1]

print("============================================================")
print(" 1. FEATURE IMPORTANCE (FAKTOR DOMINAN PENYEBAB STRES)")
print("============================================================")
for rank, idx in enumerate(indices):
    print(
        f"Peringkat {rank+1}: {X.columns[idx]} (Bobot Pengaruh: {importances[idx]*100:.2f}%)"
    )

# Plot Grafik Batang Feature Importance
plt.figure(figsize=(10, 5))
plt.title("Faktor Dominan Penyebab Stres (Feature Importance)")
plt.bar(
    range(X.shape[1]), importances[indices], color="teal", align="center"
)
plt.xticks(range(X.shape[1]), [X.columns[i] for i in indices], rotation=45)
plt.tight_layout()
plt.show()

# ============================================================
# B. STRUKTUR POHON KEPUTUSAN (VISUAL)
# ============================================================
target_names = ["Rendah (0)", "Sedang (1)", "Tinggi (2)"]
plt.figure(figsize=(20, 10))
plot_tree(
    dt_model,
    feature_names=list(X.columns),
    class_names=target_names,
    filled=True,
    rounded=True,
    fontsize=10,
)
plt.title("Struktur Pohon Keputusan (Max Depth = 3 untuk Interpretability)")
plt.show()

# ============================================================
# C. ATURAN KLASIFIKASI UTAMA (TEXT RULES)
# ============================================================
print("\n============================================================")
print(" 2. ATURAN KLASIFIKASI UTAMA (DECISION RULES)")
print("============================================================")
tree_rules = export_text(
    dt_model, feature_names=list(X.columns), show_weights=True
)
print(tree_rules)
```

## 4. Error Analisis

**1.Decision Tree (DT)**

Analisis Karakteristik: Berpotensi mengalami overfitting dan salah memprediksi pada data yang berada di batas kelas karena sifat aturan pemisahannya yang kaku.

Contoh : Ditemukan kesalahan klasifikasi pada data mahasiswa dengan kondisi fisik normal (misalnya nilai blood pressure dan heart rate stabil di angka rata-rata), namun memiliki nilai anxiety level (tingkat kecemasan) dan study load yang sangat tinggi.

Analisis Kesalahan: Model DT gagal karena di awal pohon ia membuat aturan kaku berbasis indikator fisik utama. Akibatnya, mahasiswa yang secara psikologis mengalami stres berat akibat beban kuliah tetapi fisiknya masih terlihat bugar, tetap dipaksa masuk ke label "Stres Rendah" oleh model.

**2.Naïve Bayes (NB)**

Analisis Karakteristik: Naïve Bayes mengasumsikan setiap fitur bekerja secara terpisah (independen), padahal pada data tingkat stres beberapa faktor saling berkaitan erat.

Contoh : Terdapat data mahasiswa yang memiliki skor sleep quality (kualitas tidur) buruk dan sering mengalami headache (sakit kepala), namun indikator kenyamanan lingkungan dan sosialnya sangat baik. Model NB memprediksi mahasiswa ini berada pada kategori "Stres Tinggi", padahal label aslinya adalah "Stres Sedang".

Analisis Kesalahan: Karena menganggap fitur berdiri sendiri, NB langsung menjumlahkan probabilitas tinggi dari efek kurang tidur dan sakit kepala tanpa mempertimbangkan bahwa faktor dukungan sosial yang baik bisa meredam tingkat stres tersebut di dunia nyata.

**3.K-Nearest Neighbors (KNN)**

Analisis Karakteristik: Sangat sensitif terhadap tetangga di area yang memiliki campuran kelas (area abu-abu/perbatasan).

Contoh : Ditemukan kegagalan prediksi pada data mahasiswa yang memiliki nilai fitur serba di tengah-tengah (misalkan nilai anxiety, depression, dan peer pressure berada di angka moderat). Model KNN memprediksinya sebagai "Stres Tinggi" padahal aslinya "Stres Sedang".

Analisis Kesalahan: Ketika posisi data mahasiswa tersebut berada tepat di perbatasan, KNN mengambil keputusan murni berdasarkan mayoritas label dari K tetangga terdekatnya. Karena data di area perbatasan ini sangat rapat dan bercampur, model akhirnya "terkecoh" oleh beberapa data tetangga berlabel Stres Tinggi yang kebetulan posisinya sedikit lebih dekat secara jarak Euclidean.
