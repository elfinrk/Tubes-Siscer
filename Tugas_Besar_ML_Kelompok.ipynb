### 1.1 Latar Belakang
Stres merupakan salah satu permasalahan yang banyak dialami mahasiswa selama menjalani perkuliahan. Tingginya tuntutan akademik dapat memengaruhi kondisi psikologis mahasiswa dan berdampak pada penurunan konsentrasi maupun performa belajar [1], [2]. Berdasarkan temuan tersebut, stres pada mahasiswa menjadi permasalahan yang penting untuk diperhatikan karena dapat memengaruhi proses pembelajaran dan kesejahteraan mahasiswa.

Tingkat stres mahasiswa dipengaruhi oleh berbagai faktor, di antaranya faktor akademik dan gaya hidup. Faktor akademik meliputi performa akademik, beban belajar, hubungan antara mahasiswa dan dosen, serta kekhawatiran terhadap karier di masa depan [2], [3]. Selain itu, kualitas tidur dan aktivitas mahasiswa di luar kegiatan akademik juga diketahui berhubungan dengan tingkat stres yang dialami mahasiswa [4], [5]. Oleh karena itu, faktor akademik dan gaya hidup dipilih sebagai fokus penelitian untuk mengidentifikasi tingkat stres mahasiswa.

Perkembangan Machine Learning memungkinkan proses klasifikasi data dilakukan secara otomatis berdasarkan pola yang terdapat pada data [6]. Berbagai algoritma klasifikasi telah digunakan dalam penelitian sebelumnya, termasuk Decision Tree dan K-Nearest Neighbors (KNN), yang dikenal memiliki implementasi sederhana dan performa yang cukup baik dalam tugas klasifikasi [7], [8]. Berdasarkan hal tersebut, penelitian ini bertujuan untuk membandingkan kinerja algoritma Decision Tree dan K-Nearest Neighbors dalam mengklasifikasikan tingkat stres mahasiswa berdasarkan faktor akademik dan gaya hidup menggunakan dataset Student Stress Factors: A Comprehensive Analysis.

Deskripsi dataset

Tujuan utama dataset ini adalah untuk memberikan pandangan holistik mengenai faktor-faktor yang memengaruhi kesejahteraan mental siswa. Data ini mencakup berbagai variabel dari survei nasional yang memungkinkan eksplorasi mendalam tentang bagaimana lingkungan, akademis, dan faktor personal berkontribusi terhadap tingkat stres.

Dataset ini memiliki sekitar 20 fitur yang memberikan pengaruh terbesar terhadap tingkat stres mahasiswa. Fitur-fitur tersebut dipilih secara ilmiah berdasarkan 5 faktor utama, yaitu:

Faktor Psikologis
* anxiety_level (tingkat kecemasan)
* self_esteem (harga diri)
* mental_health_history (riwayat kesehatan mental)
* depression (depresi)
  
Faktor Fisiologis
* headache (sakit kepala)
* blood_pressure (tekanan darah)
* sleep_quality (kualitas tidur)
* breathing_problem (masalah pernapasan)
  
Faktor Lingkungan
* noise_level (tingkat kebisingan)
* living_conditions (kondisi tempat tinggal)
* safety (keamanan)
* basic_needs (kebutuhan dasar)
  
Faktor Akademik
* academic_performance (prestasi akademik)
* study_load (beban belajar)
* teacher_student_relationship (hubungan dosen/guru dan mahasiswa)
* future_career_concerns (kekhawatiran terhadap karier masa depan)
  
Faktor Sosial
* social_support (dukungan sosial)
* peer_pressure (tekanan dari teman sebaya)
* extracurricular_activities (kegiatan ekstrakurikuler)
* bullying (perundungan)
