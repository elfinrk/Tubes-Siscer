#### 1.1 Latar Belakang
Stres merupakan salah satu permasalahan psikologis yang dapat memengaruhi kondisi psikologis, kesehatan, serta produktivitas individu. Penelitian ini menggunakan dataset Student Stress Factors: A Comprehensive Analysis yang mencakup faktor akademik, psikologis, sosial, fisiologis, dan lingkungan [4]. Penggunaan nilai absolut dilakukan agar fitur dengan hubungan negatif tetap dipertahankan karena tetap menunjukkan hubungan yang kuat terhadap stress_level, meskipun memiliki arah hubungan yang berlawanan. ## 2. Pra-Pemrosesan Data

#### 2.1 Pengecekan Missing Values

Konteks : Cell ini digunakan untuk mendeteksi keberadaan data kosong di setiap kolom lewat df.isnull().sum(), serta menghitung total keseluruhan data yang bolong di seluruh tabel secara instan menggunakan df.isnull().sum().sum().

- `df.isnull()`: Berfungsi untuk mengecek setiap kotak data di tabel, jika datanya kosong (null) akan ditandai True.

- `.sum() pertama`: Menjumlahkan data yang kosong di tiap-tiap kolom (misal: kolom tidur kosong 0, kolom cemas kosong 0).

- `.sum() kedua:` Menjumlahkan kembali seluruh hasil kolom tadi menjadi satu angka tunggal. Jika hasilnya 0, artinya dataset beneran bersih total dan tidak ada satu pun data yang kosong di seluruh tabel.

```python
print("Pengecekan Data Kosong")
print(df.isnull().sum())

print("\nTotal Data Kosong Dari Dataset")
print(df.isnull().sum().sum())
```

#### 2.2 Distribusi Data

Konteks : Cell ini berfungsi untuk menghitung dan menampilkan jumlah sebaran data tiap kategori stres lewat df["stress_level"].value_counts(), serta melihat persentasenya secara instan menggunakan parameter normalize=True guna memastikan keseimbangan kelas target.

- `df["stress_level"].value_counts()` : Perintah untuk menghitung total baris atau jumlah mahasiswa yang masuk ke masing-masing tingkatan stres (misal: berapa orang yang tingkat stresnya rendah, sedang, dan tinggi).

- `normalize=True`: Mengubah hitungan jumlah angka tadi menjadi bentuk desimal/proporsi (skala 0 sampai 1).

- `* 100`: Mengalikan hasil desimal dengan angka 100 agar berubah menjadi format persentase (%) yang mudah dibaca.

```python
print("Distribusi Data Per Spesies")
print(df["stress_level"].value_counts())

print("\nPersentase Distribusi Kelass (%)")
print(df["stress_level"].value_counts(normalize=True) * 100)
```

#### 2.3 Analisis Implikasi Distribusi Kelas Target terhadap Pemodelan

Berdasarkan visualisasi sebaran kelas pada variabel target stress_level (0 = Rendah, 1 = Sedang, 2 = Tinggi) yang menunjukkan jumlah sampel yang sama rata di setiap kategori, diperoleh analisis implikasi sebagai berikut:

- `Peniadaan Modifikasi Sampel`: Karena jumlah data antar-kelas seimbang secara alami, model tidak akan mengalami bias kelas mayoritas (majority class bias). Hal ini mengonfirmasi secara ilmiah bahwa kita tidak memerlukan algoritma penyeimbang data tambahan seperti SMOTE atau under-sampling.

- `Keandalan Metrik Evaluasi`: Distribusi yang seimbang ini memastikan bahwa metrik performa standar seperti Akurasi, Precision, Recall, dan F1-Score dapat dipercaya sepenuhnya secara langsung untuk mengukur kemampuan murni algoritma tanpa perlu khawatir terhadap manipulasi angka akibat data timpang.

- `Justifikasi Stratified Sampling`: Walaupun sejak awal data sudah seimbang, penggunaan skema Stratified K-Fold dalam proses validasi silang tetap krusial untuk mengunci dan memastikan bahwa rasio seimbang ini dipertahankan secara konsisten di setiap lipatan (fold) data latih dan data uji internal.
